```python
# save as export_mongo_jsonl.py
from pymongo import MongoClient, errors
from bson import json_util
import json
import sys

MONGO_URI = "mongodb+srv://ae:m0GCmWxxpFLwbqfx@tomtat.3ptutix.mongodb.net/traffic?retryWrites=true&w=majority&appName=tomtat"
DB_NAME = "traffic"          
COLLECTION_NAME = "articles" 


OUTPUT_FILE = "data_TL_check.jsonl"

client = MongoClient(MONGO_URI, serverSelectionTimeoutMS=5000)

try:
    
    client.admin.command("ping")
except errors.OperationFailure as e:
    sys.exit(1)
except errors.ServerSelectionTimeoutError as e:
    sys.exit(1)
except Exception as e:
    sys.exit(1)


db = client.get_database(DB_NAME)

try:
    cols = db.list_collection_names()
    print("Collections in DB:", cols)
except Exception as e:
    print("Không thể liệt kê collections:", e)
    cols = []

if COLLECTION_NAME not in cols:
    print(f"{COLLECTION_NAME}")
count = 0
try:
    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        cursor = db[COLLECTION_NAME].find({})
        for doc in cursor:
       
            line = json.dumps(doc, default=json_util.default, ensure_ascii=False)
            f.write(line + "\n")
            count += 1
except Exception as e:
    print("Lỗi khi ghi file hoặc khi query cursor:", e)
    sys.exit(1)

print(f"Đã xuất {count} documents từ '{COLLECTION_NAME}' ra file: {OUTPUT_FILE}")
```
